```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def mse_from_e(residual):
    #ACT-1
    return np.mean(residual ** 2) # Getting the mean of squared error vector e


# Mean Squared Error of X.w - y
def mean_squared_error(X, y, w):
    #ACT-2
    residual = X@w - y # getting the y_hat by taking the dot product of X and w
    return mse_from_e(residual) # called the above function

# The following lines are for assert statements
residual = np.array([1, 2, 3])
X = np.array([[1, 2], [3, 4]])
y = np.array([5, 6])
w = np.array([0.5, 0.5])

assert mse_from_e(residual) == np.mean(residual ** 2)
assert mean_squared_error(X, y, w) == mse_from_e(X @ w - y)
```

```python
# Generate a sawtooth weight vector
def genu(d, m):
    u = np.arange(d) * (2 * (np.arange(d) % 2) - 1)
    # u = m * u / np.sqrt(np.dot(u, u))
    return (m * u / np.sqrt(np.dot(u, u)))

# Generate random data
def genx(n, d):
    X = np.random.randint(0, 2, (n, d))
    X[:,int(d/2)] = 1
    return X

# Generate targets and add noise
def gent(X, u, noise):
    n = X.shape[0]
    y = np.dot(X, u).reshape(n, 1)
    y += noise * np.var(y) * np.random.randn(n, 1)
    return y

# Generate data, weights, and targets
def gimme_data_regres(n, d, noise=0.1):
    u = genu(d, 1.0)
    X = genx(n, d)
    y = gent(X, u, noise)
    mse_gen = mean_squared_error(X, y, u)
    print('Generator Loss={0:8.5f}\n'.format(mse_gen))
    return X, u, y
```

```python
# Plot loss as a function of epoch
use_notebook = False

def loss_plotter(vlist, fname):
    vr = vlist[0]
    vn = vlist[1]
    if not use_notebook:
        plt.figure(fname)
    plt.plot(range(1, 1+len(vr)), vr,
           range(1, 1+len(vn)), vn,
           linewidth=2, linestyle='-', marker='o')
    plt.legend(('rep', 'nor'))
    plt.grid()
    xt = np.arange(1, 1 + max(len(vr), len(vn)))
    _ = plt.xticks(xt)
    _ = plt.xlabel('Epoch', fontsize=14)
    _ = plt.ylabel(fname, fontsize=14)
    if not use_notebook:
        plt.show(block=False)
    return

# Scatter plot of predicted vs. observed targets
def loss_scatter(X, y, w, fname):
    if not use_notebook:
        plt.figure(fname)
    plt.scatter(y, X.dot(w), edgecolors=(0,0,0))
    plt.plot([y.min(), y.max()], [y.min(), y.max()], 'k--', lw=4)
    plt.grid()
    plt.xlabel('$y$', fontsize=14)
    plt.ylabel('$\hat{y}$', fontsize=14)
    if not use_notebook:
        plt.show(block=False)
```

```python
# sample new index (w/ or w/o replacement)
def sample_new_index(d, replace=1):
    if replace:
        ## ACT3
        ind = np.random.randint(0, d)

    else:
        if 'prm' not in sample_new_index.__dict__:
            sample_new_index.prm = np.random.permutation(d)
            sample_new_index.head = 0

        ## ACT4
        ind = sample_new_index.prm[sample_new_index.head]
        sample_new_index.head += 1

        if sample_new_index.head == d:
            sample_new_index.head = 0
            del sample_new_index.prm
    return ind

# The following lines are for assert statements
# Check is a boolean variable, make it true if assert statements are required.
# I am getting an error in the last chunk of this notebook when replace=0, so I kept it under the condition
check = False
if check:
    d1 = 10
    ind1 = sample_new_index(d1, replace=1)
    ind2 = sample_new_index(d1, replace=0)
    assert 0 <= ind1 < d1
    assert 0 <= ind2 < d1
```

```python
# calculate the change to w[j] wrt current margins z
# xjs is the squared norm of the jth column of X, a.k.a. ||xj||^2
def delta_wj(e, xj, xjs):
    ## ACT5
    return -(e.T @ xj)/xjs # Taking the transpose of error vector and dot product with each row of X matrix.


# Return new values for w[j] and residual
def update(wj, e, xj, xjs):
    ## ACT6
    d = delta_wj(e, xj, xjs)
    return (wj + d), (e + d * xj)


# The following lines are for assert statements
e = np.array([1, 2, 3])
xj = np.array([4, 5, 6])
xjs = np.sum(xj ** 2)

wj2 = np.array([1, 2, 3])
e2 = np.array([4, 5, 6])
xj2 = np.array([7, 8, 9])
xjs2 = np.sum(xj ** 2)
new_wj2, new_residual2 = update(wj2, e2, xj2, xjs2)

assert delta_wj(e, xj, xjs) == -(e @ xj) / xjs
assert np.all(new_wj2 == (wj2 - ((e2 @ xj2) / xjs2)))
assert np.all(new_residual2 == (e2 - (e2 @ xj2) / xjs2 * xj2))
```

```python
# Initialize all variables using the zero vector for w
# (Initialize w as the zero vector)
# You should return w, xjs, residual
def initialize(X, y):
    # ACT7
    w = np.zeros(X.shape[1])
    xjs = np.sum(X ** 2, axis=0)
    residual = -y
    return w, xjs, residual


# The following lines are for assert statements
X = np.array([[1, 2], [3, 4]])
y = np.array([5, 6])
w, xjs, residual = initialize(X, y)
assert np.all(w == np.zeros(X.shape[1]))
assert np.all(xjs == np.sum(X ** 2, axis=0))
assert np.all(residual == -y)
```

```python
# Check whether termination condition is met
def mse_check(mse_p, mse_c, eps):
    """
    This functions gives the normalized difference between mse_p and mse_c, whether it is less than the epsilon given.
    """
    ## ACT8
    return abs(mse_c - mse_p)/mse_c <= eps

mse_p = 10.0
mse_c = 9.0
eps = 0.2
assert (mse_check(mse_p, mse_c, eps))
```

```python
# Linear regression using coordinate decent
def linear_regression_cd(X, y, epochs=20, eps=0.001, replace=1):
    w, xjs, residual = initialize(X, y)
    mse_cd = [mse_from_e(residual)]
    n, d = X.shape
    for e in range(d * epochs):
        j = sample_new_index(d, replace)
        xj = X[:,j].reshape(n, 1)
        w[j], residual = update(w[j], residual, xj, xjs[j])
        if (e + 1) % d == 0:
            mse_cd.append(mse_from_e(residual))
            print('Epoch: {0:2d}  MSE: {1:5.3f}'.format(int((e+1)/d), mse_cd[-1]))
            if mse_check(mse_cd[-2], mse_cd[-1], eps): break
    return w, mse_cd
```

```python
# ---------------- Main for linear regression using Coordinate Descent --------------

np.random.seed(17)
n, d, noise = 1000, 20, 1.0
myeps = 1e-4

[X, u, y] = gimme_data_regres(n, d, noise)

mse_list = []
[wr, mse_r] = linear_regression_cd(X, y, eps=myeps)
mse_list.append(mse_r)
[wn, mse_n] = linear_regression_cd(X, y, eps=myeps, replace=0)
mse_list.append(mse_n)

if not use_notebook:
    plt.close('all')
loss_plotter(mse_list, 'MSE')
loss_scatter(X, y, wn, 'True vs. Predicted Outcome')
```
